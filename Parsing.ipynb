```python
import spacy

nlp = spacy.load("en_core_web_sm")
```

```python
#functions to find most often appearing parts of speech etc.
from operator import itemgetter
def most_pos(df, N):
    POS_dict = {}
    for text in df['Transcript']:
        doc = nlp(text)
        for token in doc:
            pos = token.pos_
            if pos in POS_dict.keys():
                POS_dict[pos] += 1
            else:
                POS_dict[pos] = 1
    final_dict = dict(sorted(POS_dict.items(), key = itemgetter(1), reverse=True)[:N])
    return final_dict

def most_tag(df, N):
    word_dict = {}
    for text in df['Transcript']:
        doc = nlp(text)
        for token in doc:
            pos = token.tag_
            if pos in word_dict.keys():
                word_dict[pos] += 1
            else:
                word_dict[pos] = 1
    final_dict = dict(sorted(word_dict.items(), key = itemgetter(1), reverse=True)[:N])
    return final_dict

def most_dep(df, N):
    word_dict = {}
    for text in df['Transcript']:
        doc = nlp(text)
        for token in doc:
            pos = token.dep_
            if pos in word_dict.keys():
                word_dict[pos] += 1
            else:
                word_dict[pos] = 1
    final_dict = dict(sorted(word_dict.items(), key = itemgetter(1), reverse=True)[:N])
    return final_dict

def most_stop(df, N):
    word_dict = {}
    for text in df['Transcript']:
        doc = nlp(text)
        for token in doc:
            pos = token.is_stop
            if pos in word_dict.keys():
                word_dict[pos] += 1
            else:
                word_dict[pos] = 1
    final_dict = dict(sorted(word_dict.items(), key = itemgetter(1), reverse=True)[:N])
    return final_dict

def most_lemma(df, N):
    word_dict = {}
    for text in df['Transcript']:
        doc = nlp(text)
        for token in doc:
            pos = token.lemma_
            if pos in word_dict.keys():
                word_dict[pos] += 1
            else:
                word_dict[pos] = 1
    final_dict = dict(sorted(word_dict.items(), key = itemgetter(1), reverse=True)[:N])
    return final_dict
```

```python
print(most_pos(data, 5))
print(most_tag(data, 5))
print(most_dep(data, 5))
print(most_stop(data, 2))
print(most_lemma(data, 5))
```

```python
#combination of title and subtitle, but also includes publishing info, author, etc.
#for relative frequencies
def len_cover(df):
    len_list = []
    for title in df['Transcript']:
        wordList = title.split()
        len_list.append(len(wordList))
    df['Words in Whole Cover'] = len_list

len_cover(data)
data.tail()
```

```python
import spacy

parsed_df = pd.DataFrame(columns = ['Date', 'Most POS', 'Most TAG', 'Most DEP', 'Num of Stopwords', 'Most Lemma', 'Total Words'])
nlp = spacy.load("en_core_web_sm")
for a in range(10, 330, 10):
    df1 = data.iloc[a-10:a]
    total = df1.sum(axis=0)[5]
    print(total)
    pos = most_pos(df1, 5)
    tag = most_tag(df1, 5)
    dep = most_dep(df1, 5)
    stop = most_stop(df1, 2)
    lemma = most_lemma(df1, 5)
    period = 1700 + (a-10)
    df2 = pd.DataFrame({'Date': period, 'Most POS': [pos], 'Most TAG': [tag], 'Most DEP':[dep], 'Num of Stopwords': [stop], 'Most Lemma':[lemma], 'Total Words':total})
    parsed_df = pd.concat([parsed_df, df2], ignore_index = True)      
```
